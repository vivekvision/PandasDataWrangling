```python
df = pd.read_excel("./MultiDimensionalSample.xlsx", sheet_name="Sheet1", header=[0,1,2,3], index_col=[0,1,2,3], dtype=str)
```

```python
df.columns = df.columns.map('_'.join)
```

```python
# reset the index 
df.reset_index(drop=False,inplace=True)
```

```python
#  rename column 1, 2, 3 & 4
```

```python
df.rename(columns={'level_0':'Region', 'level_1':'Country', 'level_2':'Sub', 'level_3':'Item'}, inplace=True)
```

```python
df.head(4)
```

```python
df = df.melt(id_vars=['Region', 'Country', 'Sub', 'Item'])
```

```python
df.head(3)
```

```python
df['column_split'] = df['variable'].str.split('_') 
```

```python
df.head(3)
```

```python
df['Year'] = df['column_split'].str[0]
df['Month'] = df['column_split'].str[1]
df['Metrics'] = df['column_split'].str[2]
df['ComulativeFreq'] = df['column_split'].str[3]
```

```python
df = df.drop('column_split', axis=1)
df = df.drop('variable', axis=1)
```
