```python
import pandas as pd

#create DataFrame
employee_df = pd.DataFrame({'id': [1,2,3,4,5,6,7],
                   'name': ['Joe', 'Henry', 'Sam', 'Max', 'Janet', 'Randy', 'Will'],
                   'salary' : [85000, 80000, 60000, 90000, 69000, 85000, 70000],
                    'departmentId' : [1,2,2,1,1,1,1]
                  })
```

```python
#create DataFrame
dept_df = pd.DataFrame({'dep_id': [1,2],
                   'dep_name': ['IT', 'Sales']
                  })
```

```python
consolidated_df = pd.merge(employee_df, dept_df, how="left",\
                           left_on=['departmentId'],\
                           right_on=['dep_id'])
```

```python
consolidated_df
```

```python
consolidated_df['salary_rank'] =\
consolidated_df.groupby(['dep_name'])['salary'].rank(method='dense',\
                                                 ascending=False)
```

```python
consolidated_df.loc[consolidated_df['salary_rank']<=3]
```
