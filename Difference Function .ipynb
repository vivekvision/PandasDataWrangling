# Forward fill

```python
df = df.ffill(axis=0)
```

# Difference Function 

```python
df['DailyOccupancy'] = df['CommutativeOccupancy'].diff()
```

```python
df['DailyOccupancy'] = df['DailyOccupancy'].fillna(df['CommutativeOccupancy'])
```
