```python
import numpy as np
import pandas as pd
```

```python
N = 10
df = pd.DataFrame({
    'type': np.random.randint(10, size=N),
    'status': np.random.randint(10, size=N),
    'name': np.random.randint(10, size=N),
    'score': np.random.randint(10, size=N),
})
```

```python
df
```

# Aggregate an existing column 'score' with lambda function

```python
df = df.groupby(['type', 'status', 'name'], as_index=False).agg({'score': lambda x: 100* (x.sum() * x.mean())})
```
