```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Math': [90, 85, 95],
    'Science': [80, 75, 85],
    'English': [95, 90, 85],
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Reshape the DataFrame using melt
melted_df = pd.melt(df, id_vars=['Name'], var_name='Subject', value_name='Grade')

melted_df
```
