```python
import pandas as pd

# Create a sample multi-index DataFrame
data = {
    ('Alice', 'Math'): [90, 85, 95],
    ('Alice', 'Science'): [80, 75, 85],
    ('Bob', 'Math'): [95, 90, 85],
    ('Bob', 'Science'): [85, 80, 75],
    ('Charlie', 'Math'): [80, 75, 85],
    ('Charlie', 'Science'): [90, 85, 95],
}
df = pd.DataFrame(data, index=['Q1', 'Q2', 'Q3'])
```

```python
df
```

```python
df.reset_index()
```

```python
# Reshape the multi-index DataFrame using melt
melted_df = pd.melt(df.reset_index(), id_vars=['index'], var_name=['Name', 'Subject'], value_name='Grade')
```
