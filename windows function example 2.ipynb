```python
import pandas as pd
import numpy as np

# create a sample dataframe
df = pd.DataFrame({'value': np.random.randint(0, 100, 20)})
```

```python
df
```

```python
# create a column to represent rolling average
window_size = 3
df['rolling_avg'] = df['value'].rolling(window_size, min_periods=1).mean()

# create a column to represent rolling sum of every other value
window_size = 4
df['rolling_sum'] = df['value'].rolling(window_size, min_periods=1).apply(
    lambda x: np.sum(x[::2])
)

# create a column to represent rolling max of every 3rd value
window_size = 5
df['rolling_max'] = df['value'].rolling(window_size, min_periods=1).apply(
    lambda x: np.max(x[::3])
)


df
```
